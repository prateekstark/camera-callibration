```python
import sys
sys.path.remove('/opt/ros/kinetic/lib/python2.7/dist-packages')
import numpy as np
import cv2
from matplotlib import pyplot
```

```python
image1 = cv2.imread('left1.16mm.jpg')
image2 = cv2.imread('left2.16mm.jpg')
```

```python
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
gray1 = np.float32(gray1)
gray2 = np.float32(gray2)

corners1 = cv2.goodFeaturesToTrack(gray1, 100, 0.01, 10)
corners1 = np.int0(corners1)
first_corner = []
second_corner = []
for corner in corners1:
    x,y = corner.ravel()
    first_corner.append([x,y])

corners2 = cv2.goodFeaturesToTrack(gray2, 100, 0.01, 10)
corners2 = np.int0(corners2)
second_corner = []
for corner in corners2:
    x,y = corner.ravel()
    second_corner.append([x,y])
    
first_corner.sort(key = lambda x: x[1])
second_corner.sort(key = lambda x: x[1])
first_point = 13
second_point = 14
cv2.circle(image1,(first_corner[first_point][0],first_corner[first_point][1]),3,255,-1)
cv2.circle(image2,(second_corner[second_point][0],second_corner[second_point][1]),3,255,-1)
```
